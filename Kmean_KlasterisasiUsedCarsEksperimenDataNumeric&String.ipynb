```python
#inisialisasi
np.random.seed(500) 
k = 3
```

```python
# inisialisasi centorid
centroids = {
    i+1: [np.random.choice(df['odometer']), np.random.choice(df['title_status'])]
    for i in range(k)
}                        
fig = plt.figure(figsize=(5, 5))
plt.scatter(df['odometer'], df['title_status'], color='k')
colmap = {1: 'r', 2: 'g', 3: 'b'}
for i in centroids.keys():
    plt.scatter(*centroids[i], color=colmap[i])
plt.show()
```

```python
#Pembuatan cluster dengan perhitungan centroid terdekat

def assignment(df, centroids):
    for i in centroids.keys():
        # sqrt((x1 - x2)^2 - (y1 - y2)^2)
        df['distance_from_{}'.format(i)] = (
            np.sqrt(
                (df['odometer'] - centroids[i][0]) ** 2
                + (df['title_status'] - centroids[i][1]) ** 2
            )
        )
    centroid_distance_cols = ['distance_from_{}'.format(i) for i in centroids.keys()]
    df['closest'] = df.loc[:, centroid_distance_cols].idxmin(axis=1)
    df['closest'] = df['closest'].map(lambda year: int(year.lstrip('distance_from_')))
    df['color'] = df['closest'].map(lambda year: colmap [year])
    return df

df = assignment(df, centroids)
print(df.head())

fig = plt.figure(figsize=(5, 5))
plt.scatter(df['odometer'], df['title_status'], color=df['color'], alpha=0.5, edgecolor='k')
for i in centroids.keys():
    plt.scatter(*centroids[i], color=colmap[i])
plt.show()
```

```python
#perbarui centorid menjadi rata - rata baru dan cluster baru

import copy

old_centroids = copy.deepcopy(centroids)

def update(k):
    for i in centroids.keys():
        centroids[i][0] = np.mean(df[df['closest'] == i]['odometer'])
        centroids[i][1] = np.mean(df[df['closest'] == i]['title_status'])
    return k

centroids = update(centroids)

fig = plt.figure(figsize=(5, 5))
ax = plt.axes()
plt.scatter(df['odometer'], df['title_status'], color=df['color'], alpha=0.5, edgecolor='k')

for i in centroids.keys():
    plt.scatter(*centroids[i], color=colmap[i])
    
for i in old_centroids.keys():
    old_year = old_centroids[i][0]
    old_odometer = old_centroids[i][1]
    dyear = (centroids[i][0] - old_centroids[i][0]) * 0.75
    dodometer = (centroids[i][1] - old_centroids[i][1]) * 0.75
    ax.arrow(old_year, old_odometer, dyear, dodometer, head_width=2, head_length=3, fc=colmap[i], ec=colmap[i])
plt.show()    
```

```python
# ulangi perbaruan centroid menjadi cluste

df = assignment(df, centroids)

fig = plt.figure(figsize=(5, 5))
plt.scatter(df['odometer'], df['title_status'], color=df['color'], alpha=0.5, edgecolor='k')
for i in centroids.keys():
    plt.scatter(*centroids[i], color=colmap[i])
plt.show()
```

```python
#memastikan sampai semua kategorisasi tidak berubah lagi

while True:
    closest_centroids = df['closest'].copy(deep=True)
    centroids = update(centroids)
    df = assignment(df,centroids)
    if closest_centroids.equals(df['closest']):
        break
    
fig = plt.figure(figsize=(5, 5))
plt.scatter(df['odometer'], df['title_status'], color=df['color'], alpha=0.5, edgecolor='k')
for i in centroids.keys():
    plt.scatter(*centroids[i], color=colmap[i])

plt.show()
```
